```python
# Module one covers Binary Search,Linked Lists and Time Complexity
from jovian.pythondsa import evaluate_test_cases
```

```python
sample_tests = []
# query occurs in the middle
sample_tests.append({
    'input': {
        'cards': [13, 11, 10, 7, 4, 3, 1, 0],
        'query': 1
    },
    'output': 6
})
# query is the first element
sample_tests.append({
    'input': {
        'cards': [4, 2, 1, -1],
        'query': 4
    },
    'output': 0
})
# query is the last element
sample_tests.append({
    'input': {
        'cards': [3, -1, -9, -127],
        'query': -127
    },
    'output': 3
})
# cards contains just one element, query
sample_tests.append({
    'input': {
        'cards': [6],
        'query': 6
    },
    'output': 0 
})
# cards does not contain query 
sample_tests.append({
    'input': {
        'cards': [9, 7, 5, 2, -9],
        'query': 4
    },
    'output': -1
})
# cards is empty
sample_tests.append({
    'input': {
        'cards': [],
        'query': 7
    },
    'output': -1
})
# numbers can repeat in cards
sample_tests.append({
    'input': {
        'cards': [8, 8, 6, 6, 6, 6, 6, 3, 2, 2, 2, 0, 0, 0],
        'query': 3
    },
    'output': 7
})
# query occurs multiple times
sample_tests.append({
    'input': {
        'cards': [8, 8, 6, 6, 6, 6, 6, 6, 3, 2, 2, 2, 0, 0, 0],
        'query': 6
    },
    'output': 2
})
# query occurs in the middle
sample_tests.append({
    'input': {
        'cards': [13, 11, 10, 7, 4, 3, 1, 0],
        'query': 1
    },
    'output': 6
})
```

```python
#Linear search approach
def locate_card(cards,query):
    position = 0
    while position < len(cards):
        if cards[position] == query:
            return position
        else:
            position += 1 
    return -1

expected_result = evaluate_test_cases(locate_card,sample_tests)
print(expected_result)
```

```python
# Binary Search
def test_location(cards,query,mid):
    mid_num = cards[mid]
    if mid_num == query:
        if (mid -1 > 0) and (cards[mid - 1]):
            return 'left'
        else:
            return 'found'
    elif mid_num < query:
        return 'left'
    else:
        return 'right'
    
def find_card(cards, query):
    low_index, high_index = 0, len(cards) - 1
    while low_index <= high_index:
        mid = low_index + high_index // 2
        result = test_location(cards, query, mid)
        if result == 'found':
            return mid
        elif result == 'left':
            high_index = mid - 1
        elif result == 'right':
            low_index = mid + 1
    return -1
evaluate_test_cases(find_card,sample_tests)
```
